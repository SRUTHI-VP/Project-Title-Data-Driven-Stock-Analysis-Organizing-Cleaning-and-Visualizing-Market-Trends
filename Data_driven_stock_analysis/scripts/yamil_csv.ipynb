```python
import os
import yaml
import pandas as pd

# Root directory containing all month/date folders
root_dir = "C:/Users/LENOVO/Desktop/GUVI/Project/nifty50"
output_dir = "C:/Users/LENOVO/Desktop/GUVI/Project/data/tickers"
os.makedirs(output_dir, exist_ok=True)

# Dictionary to hold data for each stock
stock_data = {}

# Recursively walk through all subfolders to find YAML files
for dirpath, _, files in os.walk(root_dir):
    for file in files:
        if file.endswith(".yaml") or file.endswith(".yml"):
            file_path = os.path.join(dirpath, file)
            with open(file_path, "r") as f:
                data = yaml.safe_load(f)

            # Expecting a list of records
            if isinstance(data, list):
                for record in data:
                    ticker = record.get("Ticker")  # Capital T!
                    if ticker:
                        if ticker not in stock_data:
                            stock_data[ticker] = []
                        stock_data[ticker].append(record)

# Write each ticker's data to its own CSV file
for ticker, records in stock_data.items():
    df = pd.DataFrame(records)
    output_file = os.path.join(output_dir, f"{ticker}.csv")
    df.to_csv(output_file, index=False)
    print(f"âœ… Saved: {output_file}")

print("ðŸŽ‰ All 50 stock CSVs are successfully created in:", output_dir)
```

```python
import os, pandas as pd

DATA_DIR = "C:/Users/LENOVO/Desktop/GUVI/Project/data/tickers"   # folder with your 50 ticker CSVs
OUTPUT_DIR = "C:/Users/LENOVO/Desktop/GUVI/Project/output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def load_all_tickers():
    frames = []
    for file in os.listdir(DATA_DIR):
        if file.endswith(".csv"):
            df = pd.read_csv(os.path.join(DATA_DIR, file))
            df["date"] = pd.to_datetime(df["date"], errors="coerce")
            df["Ticker"] = df["Ticker"].astype(str)
            df["month"] = df["date"].dt.to_period("M").astype(str)
            frames.append(df)
    master = pd.concat(frames, ignore_index=True)
    master.sort_values(["Ticker","date"], inplace=True)
    return master

master_df = load_all_tickers()
master_df.to_csv(os.path.join(OUTPUT_DIR, "clean_master.csv"), index=False)
print("âœ… Saved clean_master.csv")
```
